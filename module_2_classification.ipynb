## Classification of IRIS Data

### Step-1 Loading Necessary libraries

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)
```

### Step-2 Loading the dataset


```python
from sklearn import datasets
# import some data to play with
iris = datasets.load_iris()
```

```python
iris
```

```python
print(iris.DESCR)
```

```python
iris.data
```

Creating data from the iris dictionary

```python
dataset = pd.DataFrame(data=iris.data, columns=["sepal_length", "sepal_width", "petal_length", "petal_width"])
dataset.head() # showing first five rows of the dataframe
```

```python
# appending the target as last column
dataset["target"] = iris.target
target_names = {0: "Iris-Setosa", 1: "Iris-Versicolour", 2:"Iris-Virginica" }
dataset.head()
```

```python
# encoding the target labels
target_names = {0: "Iris-Setosa", 1: "Iris-Versicolour", 2:"Iris-Virginica" }
dataset['target'] =dataset['target'].map(target_names)
dataset.head()
```

## Preliminary EDA on the data

First of all the class-wise balance of the data is verified using the count function. Here each class contains 50 samples each. So it is a well balanced data.

```python
# Let's see how many examples we have of each species as well as the data type for
# each species
dataset["target"].value_counts()
```

## Quality of features in separating one class from other

```python
sns.boxplot(x="target", y="sepal_length", palette="husl", data=dataset)
```

```python
sns.boxplot(x="target", y="sepal_width", palette="husl", data=dataset)
```

```python
sns.boxplot(x="target", y="petal_length", palette="husl", data=dataset)
```

```python
sns.boxplot(x="target", y="petal_width", palette="husl", data=dataset)
```
