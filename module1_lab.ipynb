Reading a vector in python


```python
import numpy as np
v1=np.array([1,2,3])
v2=np.array([2,4,5])
a=2.3
print("The vector is: {}".format(v1))
print('The scalar product of {}  with {} is : {}'.format(v1,a,a*v1))
print('The linear combination of {} and {} is :{}'.format(v1,v2,a*v1+v2))
```

Matrix operations in python



```python
M1=np.array([[1,2,3],[2,3,4],[4,5,6]])
M2=np.array([[3,2,1],[2,4,6],[5,6,7]])
print('Sum of matrices:',M1+M2)
print('Scalar product:',a*M1)
print('Element-wise product:',M1*M2)
print('Matrix Product',np.dot(M1,M2))
```
